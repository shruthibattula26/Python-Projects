```python
# Replacing multiple names of attributes to match the content in it 

column_mapping = {'name': 'Name', 'mfr': 'Maker', 'type': 'Type', 'calories': 'Calories (per serving)', 
                  'protein': 'Protein (g)', 'fat': 'Fat (g)', 'sodium': 'Sodium (mg)', 'fiber': 'Fiber (g)', 
                  'carbo': 'Carbohydrates (g)', 'sugars': 'Sugars (g)', 'potass': 'Potassium (mg)', 'vitamins': 'Vitamins (%)', 
                  'shelf': 'Display Shelf Level', 'weight': 'Weight (ounces per serving)', 'cups': 'Cups (per serving)', 'rating': 'Rating'}

df1 = df1.rename(columns = column_mapping)

df1.tail()
```

```python
# Listing count of values in the Maker column

df1["Maker"].value_counts()
```

```python
# Listing count of values in the Type column

df1["Type"].value_counts()
```

```python
# Replacing specific values in specific columns to display full forms

df1 = df1.replace({'Maker': {'A': 'American Home Food Products', 'G': 'General Mills', 'K': 'Kelloggs', 'N': 'Nabisco', 'P': 'Post', 
                             'Q': 'Quaker Oats', 'R': 'Ralston Purina'}, 
                   'Type': {'C': 'Cold', 'H': 'Hot'}})

df1.head()
```

```python
# Listing count of values in the Maker column after corrections

df1["Maker"].value_counts()
```

```python
# Listing count of values in the Type column after corrections

df1["Type"].value_counts()
```

```python
# Tpecasting Fiber and Carbohydrates column's datatype to integer as it is appropriate

df1[['Fiber (g)', 'Carbohydrates (g)']] = df1[['Fiber (g)', 'Carbohydrates (g)']].astype(int)

df1.tail()
```

```python
# Counting the number of -1 in Carbohydrates, Sugars, and Potassium columns

df1[df1 == -1].count(axis = 0)
```

```python
# Above we can see that we have 4 negative values in those columns. Hence, Replacing them with their medians

df1 = df1.replace({'Carbohydrates (g)': {-1: df1['Carbohydrates (g)'].median()}, 
                   'Sugars (g)': {-1: df1['Sugars (g)'].median()},
                  'Potassium (mg)': {-1: df1['Potassium (mg)'].median()}})
```

```python
# Ensuring there is no count of -1 data points in Carbohydrates, Sugars, and Potassium columns

df1[df1 == -1].count(axis = 0)
```

```python
# Reducing the numbers after decimal to 2 in Rating column

df1['Rating'] = df1['Rating'].round(2)

df1.head()
```

#### What are the Factors influencing the Rating of the Cereals?

```python
# Bringing all Numeric Attributes under one variable for further analysis

numerical_var = df1.select_dtypes(include = ['int64', 'int32', 'float64']).copy()
numerical_var
```

```python
# Using strip plot to understand the distribution of our dataset

plt.figure(figsize = (12, 8))
sns.stripplot(data = df1, jitter = True, size = 5, alpha = 0.7)

# Adding labels and title

plt.title('Strip Plot of All Variables in the Dataset', fontsize = 16)
plt.xlabel('Variables', fontsize = 12)
plt.ylabel('Values', fontsize = 12)
plt.xticks(rotation = 45)

# Displaying the plot

plt.show()
```

```python
# Finding the Correlation between Attributes to understand Rating 

correlation = numerical_var.corr()
correlation
```

```python
# Listing column names to plot against 'Rating'

columns = [
    'Calories (per serving)', 'Protein (g)', 'Fat (g)', 'Sodium (mg)',
    'Fiber (g)', 'Carbohydrates (g)', 'Sugars (g)', 'Potassium (mg)',
    'Vitamins (%)', 'Display Shelf Level', 'Weight (ounces per serving)', 'Cups (per serving)'
]

# Creating a 3x4 grid of subplots

fig, axes = plt.subplots(3, 4, figsize = (20, 15))

# Flattening the axes array for easier iteration

axes = axes.flatten()

# Plotting each scatter plot

for i, col in enumerate(columns):
    sns.scatterplot(x = col, y = 'Rating', data = df1, ax = axes[i])
    axes[i].set_title(f'{col} vs Rating')

# Hiding any unused axes

for j in range(len(columns), len(axes)):
    axes[j].axis('off')

# Adjusting layout

plt.tight_layout()
plt.show()
```

```python
# Heatmap

plt.figure(figsize = (10, 10), facecolor = 'w', edgecolor = 'k')
sns.heatmap(correlation, cmap = 'viridis', annot = True, fmt = '.1f', linewidths = 1, linecolor = 'black')
```

### Conclusion:

#### This analysis reveals key insights into the factors influencing cereal ratings, offering a deeper understanding of consumer preferences:

#### 1. High Protein and Fiber Lead to Higher Ratings:

#### Cereals with higher protein and fiber content tend to receive better ratings. These nutrients are often under-consumed in a typical diet, making cereals that supplement them more appealing to health-conscious consumers. By promoting a balanced diet, such cereals align with modern nutritional priorities, potentially explaining their higher ratings.

#### Implication: Manufacturers can focus on increasing the protein and fiber content in their products to cater to health-conscious consumers and improve customer satisfaction.

#### 2. High Carbohydrates, Fats, Sugars, Sodium, and Calories Result in Lower Ratings:  
#### Conversely, cereals with high levels of carbohydrates, fats, sugars, sodium, and calories are associated with lower ratings. This indicates a shift in consumer preferences toward healthier, low-calorie, and nutrient-dense options. High sugar content and excessive sodium, in particular, are seen as unhealthy and might deter health-conscious buyers.

#### Implication: To remain competitive, cereal manufacturers should consider reformulating their products by reducing sugar, sodium, and fat content while maintaining or enhancing flavor and quality. Marketing efforts should also emphasize health benefits and balanced nutrition.

#### 3. Balancing Nutrition and Taste:  
#### While ratings suggest a clear preference for healthier cereals, its important to strike a balance between nutritional value and taste. Consumers are likely to gravitate toward cereals that deliver both health benefits and enjoyable flavors.

#### 4. Market Opportunities:  
#### The findings present opportunities for product innovation in the cereal industry. For example:
#### - Launching high-protein and high-fiber cereals aimed at fitness enthusiasts.
#### - Introducing low-calorie options with reduced sugar and sodium for health-conscious families.

####

#### Final Thoughts:
#### This analysis highlights a clear consumer preference for nutrient-dense cereals over calorie-dense, sugary, or fatty options. 
#### To stay relevant and competitive, cereal manufacturers must adapt to these trends, focusing on healthier formulations that align with consumer expectations for both nutrition and taste.
#### By addressing these insights, the cereal industry can not only boost product ratings but also foster greater customer loyalty and long-term success.
