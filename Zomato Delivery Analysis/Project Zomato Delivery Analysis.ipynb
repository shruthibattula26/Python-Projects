### Handling Outliers in 'Average Cost for two'

```python
# Finding out quartiles

Q1_Average_Cost_for_two = zomato_df_copy['Average Cost for two'].quantile(0.25)
Q3_Average_Cost_for_two = zomato_df_copy['Average Cost for two'].quantile(0.75)

print(f"Q1 is: {Q1_Average_Cost_for_two}  and   Q3 is {Q3_Average_Cost_for_two}")

IQR = Q3_Average_Cost_for_two - Q1_Average_Cost_for_two

print(IQR)
```

```python
# Calculating Lower & Upper Limits

lower_lim = Q1_Average_Cost_for_two - (1.5 * IQR)
upper_lim = Q3_Average_Cost_for_two + (1.5 * IQR)

print(lower_lim)

print(upper_lim)
```

```python
# Identifying the row index for those values that are beyond the boundaries

low_bound = (zomato_df_copy['Average Cost for two'] < lower_lim)
up_bound = (zomato_df_copy['Average Cost for two'] > upper_lim)

print(low_bound.value_counts())

print("-" * 20)

print(up_bound.value_counts())
```

```python
# Diplays the index of the outliers

zomato_df_copy['Average Cost for two'][(low_bound | up_bound)]
```

```python
# Creating a new dataframe with outliers removed

zomato_df_no_outliers = zomato_df_copy[~(low_bound | up_bound)]
zomato_df_no_outliers
```

```python
# Visualizing Data - BOX Plot - to find outlilers

sns.boxplot(x=zomato_df_no_outliers['Average Cost for two'])
```

```python
# Visualizing distribution 

sns.displot(zomato_df_no_outliers['Average Cost for two'], bins = 5, kde = True)
```
